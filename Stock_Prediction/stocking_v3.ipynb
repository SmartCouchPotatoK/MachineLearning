```python
%matplotlib inline

import pandas_datareader.data as pdr
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from pandas.plotting import register_matplotlib_converters

register_matplotlib_converters()

start=datetime.datetime(2010,1,1)
end=datetime.datetime(2019,9,27)

hdhi=pdr.DataReader('267250.KS', 'yahoo', start, end)
sshi=pdr.DataReader('010140.KS', 'yahoo', start, end)
dsme=pdr.DataReader('042660.KS', 'yahoo', start, end)

hdhi['Close'].plot(figsize=(20,10), label='Hyundai')
sshi['Close'].plot(figsize=(20,10), label='Samsung')
dsme['Close'].plot(figsize=(20,10),label='Daewoo')

plt.legend( )
plt.show( )
```
