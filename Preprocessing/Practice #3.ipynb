```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('data/iris.csv')
df[:5]
```

```python
X = df.values[:,0].astype(float)
Y = df.values[:,1].astype(float)
Z = df.values[:,2].astype(float)
```

```python
fig = plt.figure()

ax = fig.add_subplot(projection='3d')

ax.scatter(X, Y, Z)
ax.set_xlabel('sepallength')
ax.set_ylabel('sepalwidth')
ax.set_zlabel('petallength')
plt.show()
```

3D Surface Plots

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
```

```python
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', antialiased=False)
wire = ax.plot_wireframe(X, Y, Z, color='green', linewidth=0.1)

fig.colorbar(surf)
plt.show()
```
