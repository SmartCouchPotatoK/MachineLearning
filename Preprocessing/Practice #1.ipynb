```python
label_to_index = {'Iris-setosa':0,
                  'Iris-versicolor':1,
                  'Iris-virginica':2}
label_to_index['Iris-versicolor']
```

```python
y_number = [label_to_index[label] for label in y]
y_number[:5]
```

```python
plt.figure(figsize=(7,7))
plt.xlabel(columns[0])
plt.ylabel(columns[1])
plt.scatter(X[:,0], X[:,1], c=y_number, cmap='tab10')
plt.show()
```

```python
dat = X[:,0]
dat = np.sort(dat)

# 0~1사이를 30개의 구간으로 나누기
f_values = np.linspace(0.0, 1.0, 30)

N = len(dat)
dat_quantile = [dat[int(f*(N-1))] for f in f_values]
```

```python
plt.scatter(f_values, dat_quantile)
plt.xlabel('f-value')
plt.ylabel(columns[0])
plt.show()
```

```python
fig, axs = plt.subplots(4, 4, figsize = (20,20))
for i in range(4):
    for j in range(4):
        axs[i, j].set(xlabel=columns[i], ylabel=columns[j])
        axs[i, j].scatter(X[:,i],X[:,j])
        
plt.show()
```

```python
fig, axs = plt.subplots(4, 4, figsize = (20,20))
for i in range(4):
    for j in range(4):
        axs[i, j].set(xlabel=columns[i], ylabel=columns[j])
        axs[i, j].scatter(X[:,i],X[:,j], c=y_number, cmap='tab10')
        
plt.show()
```
