### 작업형 제2유형

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
```

```python
X_train = pd.read_csv('X_train.csv', encoding='cp949')
X_test = pd.read_csv('X_test.csv', encoding='cp949')
y_train = pd.read_csv('y_train.csv', encoding='cp949')
print(X_train.shape, y_train.shape, X_test.shape)
```

```python
X = pd.concat([X_train, X_test])
X.shape
```

```python
pd.options.display.float_format = '{:.2f}'.format
X.describe(include='all')
```

```python
X.isnull().sum()
```

```python
# 환불한 금액이 없으면 '0'원으로 처리
X = X.fillna(0)
X
```

```python
X_hot = pd.get_dummies(X, drop_first=True)
X_hot
```

```python
from sklearn.model_selection import train_test_split
new_X_train, new_X_test = train_test_split(X_hot[:3500], test_size=0.2, random_state=42)
new_y_train, new_y_test = train_test_split(y_train['gender'], test_size=0.2, random_state=42)
print(new_X_train.shape, new_y_train.shape, new_X_test.shape, new_y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(new_X_train, new_y_train)
y_pred = model.predict_proba(new_X_test)
pd.DataFrame(y_pred)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(new_y_test, y_pred[:, 1])
```

```python
res = pd.DataFrame(X[3500:]['cust_id'])

model.fit(X_hot[:3500], y_train['gender'])
res_predict = model.predict_proba(X_hot[3500:])
res_predict = pd.DataFrame(res_predict[:, 1])

df = pd.concat([res, res_predict], ignore_index=True, axis=1)
df = df.rename(columns = {0: 'custid', 1: 'gender'})
df.to_csv('수험번호.csv', index=False)
```
