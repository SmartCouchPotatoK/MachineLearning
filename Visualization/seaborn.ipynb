```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
# Hue = 색상 (Groupby 느낌)
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='size')
```

```python
# style & hue
sns.relplot(x='total_bill', y='tip', data=tips, style='smoker', hue='size')
```

```python
# 색상을 입혀보자
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', palette='prism')
```

```python
# 색상을 입혀보자
sns.relplot(x='total_bill', y='tip', data=tips, size='size')
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
# x값이 겹치는 경우 range를 보여준다.
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', ci=None)
```

```python
# 위에는 표준편차, 아래는 분산?
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', ci='sd')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='region')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='region', style='event')
```

* relplot : 관계형그래프 (scatter, line)
* catplot : 범주형그래프 (bar, box, count..)

```python
tips.head(3)
```

```python
# cat plot
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
# marker를 겹치지 않게 그리기
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
```

```python
# box를 여러개로 만들기
sns.catplot(x='day', y='total_bill', data=tips, kind='boxen')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='point')
```

```python
# count를 못그리는 데이터다?
sns.catplot(x='day', y='total_bill', data=tips, kind='count')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar', ci=None)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm', hue='sex')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='sex', split='True')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='sex', split='True', inner='stick')
```

Joint Plot

```python
df = pd.DataFrame(np.random.randn(500,2), columns=['1', '2'])
df.head(3)
```

```python
sns.relplot(x='1', y='2', data=df)
```

```python
sns.jointplot(x='1', y='2', data=df)
```

```python
# scatter를 육각모양으로 heatmap 같이..
sns.jointplot(x='1', y='2', data=df, kind='hex')
```

```python
sns.jointplot(x='1', y='2', data=df, kind='kde')
```

```python
# scatter랑 다른점은? marker에 투명도가 생김.
sns.jointplot(x='1', y='2', data=df, kind='resid')
```

```python
sns.jointplot(x='1', y='2', data=df, kind='reg')
```
